> * In your own words, explain the differnece between *exploratory* and *explanatory* data analysis. 

> In exploratory analysis we are basically exploring data to find relationships between different variables, finding patterns or testing our hypothesis.
When the exploratory analysis is done and we have visualized the data and found some patterns or relationships, now it is time to ask why and how and that is the explanatory analysis. 
Example, In exploratory analysis, we found out from sales data that costumers are not buying products and sales have been decreasing. In Explanatory analysis, we answer why they are not buying.
